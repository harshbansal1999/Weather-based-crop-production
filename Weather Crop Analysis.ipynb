```python
#Dealing with null values
df1['low_wind']=df1['low_wind'].fillna(df1['low_wind'].mean())
df1['events']=df1['events'].fillna('Not Specified')
```

```python
df1['low_wind']=df1['low_wind'].astype(int)
```

```python
def f(x):
    x=(x-32)*0.5
    return int(x)
```

```python
#Change temperature in Farhenheit to Celcius
df1['high_temp']=df1['high_temp'].apply(f)
df1['low_temp']=df1['low_temp'].apply(f)
```

```python
df1['Temp Diff']=df1['high_temp']-df1['low_temp']
df2['Temp Diff']=df2['Max Temp']-df2['Min Temp']
```

```python
def step_gradient(points,learning_rate,m,c):
    m_slope=0
    c_slope=0
    M=len(points)
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        m_slope=(2/M)*(y-m*x-c)*c
        c_slope+=(-2/M)*(y-m*x-c)
    new_m=m-learning_rate*m_slope
    new_c=c-learning_rate*c_slope
    return new_m,new_c
```

```python
def gd(points,learning_rate,num_iteration):
    m=0
    c=0
    for i in range(num_iteration):
        m,c=step_gradient(points,learning_rate,m,c)
        print(i,"Cost",cost(points,m,c))
    return m,c
```

```python
def cost(points,m,c):
    total_cost=0
    M=len(points)
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        total_cost=(1/M)*((y-m*x-c)**2)
    return total_cost
```

```python
def run():
    learning_rate=0.004
    num_iteration=50
    m,c=gd(df_temp,learning_rate,num_iteration)
    print(m,c)
```
